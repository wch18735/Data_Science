## [P2]
### Gather tweets of 5 Twitter users and perform the same exploration as the guide

### My user list
- Donald J. Trump (@realDonaldTrump)
- Barack Obama (@BarackObama)
- Bill Gates (@BillGates)
- Dalai Lama (@DalaiLama)
- Google (@Google)

```python
# user_list
user_list = ['realDonaldTrump', 'BarackObama', 'BillGates', 'DalaiLama', 'Google']
```

```python
for screen_name in user_list:
    user_tweets = twitter_api.statuses.user_timeline (screen_name=screen_name, 
                                                  count=100)
    print(len(user_tweets))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

CountVectorList = []
Tf_IdfList = []
```

```python
for screen_name in user_list:
    print("load %s tweets..." % screen_name)
    user_tweets = twitter_api.statuses.user_timeline (screen_name=screen_name, 
                                                  count=100)

    # print ten tweet
    print("Tweets counts:", len(user_tweets))
    
    # all text from user 
    print("extracting text in tweets...")
    user_texts = [status['text'] for status in user_tweets]
    
    # CounterVectorizer
    print("feature extract using CounterVectorizer...")
    vectorizer = CountVectorizer()
    x = vectorizer.fit_transform(user_texts)
    CountVectorList.append(x)
    
    # TF-IDF Vectorizer
    print("feature extract using TF-IDF vectorizer...")
    vectorizer = TfidfVectorizer()
    x = vectorizer.fit_transform(user_texts)
    Tf_IdfList.append(x)
    
    # Word Cloud
    print("making conbined string for word cloud")
    combined_string = ', '.join(user_texts)
    wordcloud = WordCloud().generate(combined_string)
    
    # save figure
    print("saving word cloud image in %s" % "./Lab/lab2/wordcloud")
    fig = plt.figure(figsize = (12,5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.savefig("./Lab/lab2/wordcloud/%s_wordcloud.png"%screen_name, format='png')
    
    print("%s wordcloud image save successfully.." % screen_name)
    print()
```

```python
# we can check features
print("CountVector List:", end=' ')
for i in CountVectorList:
    print(i.shape, end=' ')
    
print()
    
print("TF-IDF List:", end=' ')
for i in Tf_IdfList:
    print(i.shape, end=' ')
```
